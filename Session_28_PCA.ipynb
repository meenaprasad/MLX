## Problem Statement
We have to transform iris data into 3 dimensions and plot a 3d
chart with transformed dimensions and color each data point with specific class.

```python
# import libraries.
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
# load iris dataset.
iris = datasets.load_iris()
# select all rows and columns.
x = iris.data[:, :]
y = iris.target

x_min, x_max = x[:, 1].min() - .5, x[:, 2].max() + .5
y_min, y_max = x[:, 2].min() - .5, x[:, 2].max() + .5
```

```python
print(x)
```

```python
print(y)
```

```python
plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(x[:, 1], x[:, 2], c='b', cmap=plt.cm.Set1, edgecolor='k')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal width')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
x_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(x_reduced[:, 0], x_reduced[:, 1], x_reduced[:, 2], c=y, cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("Three PCA directions")
ax.set_xlabel("1st Eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd Eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd Eigenvector")
ax.w_zaxis.set_ticklabels([])
plt.show()
```
